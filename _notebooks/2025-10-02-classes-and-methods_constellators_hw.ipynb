```python
<!DOCTYPE html>
<html lang="en">
<head>
  <meta charset="UTF-8" />
  <meta name="viewport" content="width=device-width, initial-scale=1.0"/>
  <title>OOP Homework in JavaScript</title>
  <style>
    body { font-family: Arial, sans-serif; margin: 20px; }
    .output { white-space: pre-wrap; background: #f4f4f4; padding: 10px; border-radius: 5px; }
  </style>
</head>
<body>

<h2>JavaScript OOP Homework</h2>
<div id="output" class="output"></div>

<script>
  const outputDiv = document.getElementById("output");

  function log(message) {
    outputDiv.innerText += message + "\n";
  }

  // 1) Person class
  class Person {
    constructor(name, age) {
      this.name = name;
      this.age = age;
    }

    // 2) greet method
    greet() {
      log(`Hello! My name is ${this.name} and I am ${this.age} years old.`);
    }
  }

  // 3) Book class
  class Book {
    constructor(title, author, totalPages) {
      this.title = title;
      this.author = author;
      this.totalPages = totalPages;
      this.pagesRead = 0;
    }

    // 4) read method
    read(pages) {
      if (this.pagesRead < this.totalPages) {
        const pagesToRead = Math.min(pages, this.totalPages - this.pagesRead);
        this.pagesRead += pagesToRead;
        log(`Read ${pagesToRead} pages of '${this.title}'. Total pages read: ${this.pagesRead}/${this.totalPages}`);
      } else {
        log(`'${this.title}' is already finished!`);
      }
    }

    // 5) isLongBook method
    isLongBook() {
      return this.totalPages > 300;
    }
  }

  // 6) Car class
  class Car {
    constructor(make, model, fuel) {
      this.make = make;
      this.model = model;
      this.fuel = fuel;
    }

    drive(distance) {
      const fuelNeeded = distance * 0.1;
      if (this.fuel >= fuelNeeded) {
        this.fuel -= fuelNeeded;
        log(`The ${this.make} ${this.model} drove ${distance} km. Fuel remaining: ${this.fuel.toFixed(1)} units.`);
      } else {
        log(`Warning! Cannot drive ${distance} km. Not enough fuel (${this.fuel.toFixed(1)} units left).`);
      }
    }

    refuel(amount) {
      this.fuel += amount;
      log(`Refueled by ${amount} units. Total fuel now: ${this.fuel.toFixed(1)} units.`);
    }

    static info() {
      log("Cars are vehicles.");
    }
  }

  // 8) ElectricCar class extending Car
  class ElectricCar extends Car {
    constructor(make, model, battery) {
      super(make, model, 0); // fuel set to 0 for electric cars
      this.battery = battery;
    }

    drive(distance) {
      const chargeNeeded = distance * 0.2;
      if (this.battery >= chargeNeeded) {
        this.battery -= chargeNeeded;
        log(`The ${this.make} ${this.model} (Electric) drove ${distance} km. Battery remaining: ${this.battery.toFixed(1)}%`);
      } else {
        log(`Warning! Cannot drive ${distance} km. Low battery (${this.battery.toFixed(1)}% left).`);
      }
    }

    charge(amount) {
      this.battery = Math.min(100, this.battery + amount);
      log(`Charged by ${amount} units. Battery level is now: ${this.battery.toFixed(1)}%`);
    }
  }

  // --- Object Instantiation & Execution (like main section) ---
  log("--- OOP Exercises Execution ---");

  const person1 = new Person("Jane Doe", 25);
  const book1 = new Book("The Odyssey", "Homer", 541);
  const car1 = new Car("Toyota", "Camry", 40);
  const eCar1 = new ElectricCar("Lucid", "Air", 75);

  // Q1 & Q3
  log("\n[Q1] Person Properties:");
  log(`Name: ${person1.name}, Age: ${person1.age}`);

  log("\n[Q3] Book Properties:");
  log(`Title: ${book1.title}, Author: ${book1.author}, Pages: ${book1.totalPages}`);

  // Q2
  log("\n[Q2] Person Greeting:");
  person1.greet();

  // Q5
  log("\n[Q5] Book Length Check:");
  log(`Is '${book1.title}' a long book (> 300 pages)? ${book1.isLongBook()}`);

  // Q7
  log("\n[Q7] Car Static Info:");
  Car.info();

  // Q8
  log("\n[Q8] ElectricCar Initial State:");
  log(`Electric Car: ${eCar1.make} ${eCar1.model}, Battery: ${eCar1.battery}%`);
  eCar1.charge(15);

  // Q10
  log("\n--- [Q10] Combined Object Routine ---");
  car1.drive(200);
  eCar1.charge(50);
  book1.read(150);
  book1.read(500); // reading past total pages
  person1.greet();

  log("\n-------------------------------------");
</script>

</body>
</html>
```
