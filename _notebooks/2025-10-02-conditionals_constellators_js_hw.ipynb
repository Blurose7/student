```python
%%html

<div id="outputPH2">Results will appear here:<br></div>

<script>
(() => {
    const output = document.getElementById("outputPH2");

    let snackChoice = prompt("Choose a number: 1, 2, or 3 to get a snack!");

    if (snackChoice == "1") { // Chocolate bar
        let chocolateType = prompt("You got a chocolate bar! Choose your type: milk or white?");
        if (chocolateType.toLowerCase() === "milk") {
            output.innerText += "You got a milk chocolate bar!\n";
        } else if (chocolateType.toLowerCase() === "white") {
            output.innerText += "You got a white chocolate bar!\n";
        } else {
            output.innerText += "Unknown chocolate type!\n";
        }
    } else if (snackChoice == "2") { // Chips
        output.innerText += "You got chips!\n";
    } else if (snackChoice == "3") { // Candy
        let candyColor = prompt("You got candy! Choose your color: red, green, or yellow?");
        if (candyColor.toLowerCase() === "red") {
            output.innerText += "You got red sweet candy!\n";
        } else if (candyColor.toLowerCase() === "green") {
            output.innerText += "You got sour apple candy!\n";
        } else if (candyColor.toLowerCase() === "yellow") {
            output.innerText += "You got lemon candy!\n";
        } else {
            output.innerText += "Unknown candy color!\n";
        }
    } else {
        output.innerText += "Invalid snack choice!\n";
    }

    // Challenge: 10% chance of extra cookie
    if (Math.random() < 0.1) {
        output.innerText += "Bonus treat! The machine gives you an extra cookie!,YOU ARE ONE LUCKY FELLOW\n";
    }
})();
</script>
```

```python
%%html

<div id="outputHW">Results will appear here:<br></div>

<script>

    (() => {

        const output = document.getElementById("outputHW");

        console.log("Code, code, code!");
        output.innerText += "Code, code, code!\n\n";

        // 1) Temperature
        let temperature = 85;
        if (temperature > 80) {
            output.innerText += "It’s hot outside!\n";
        } else {
            output.innerText += "It’s nice outside.\n";
        }

        // 2 & 3) Score (no user input, just a fixed example)
        let score = 92;
        if (score >= 90) {
            output.innerText += "Grade: A\n";
        } else if (score >= 80) {
            output.innerText += "Grade: B\n";
        } else if (score >= 70) {
            output.innerText += "Grade: C\n";
        } else {
            output.innerText += "Needs improvement.\n";
        }

        // 4 & 5) Random number from 1–10, check even or odd
        let number = Math.floor(Math.random() * 10) + 1;
        if (number % 2 === 0) {
            output.innerText += `Number: ${number} — Even Number\n`;
        } else {
            output.innerText += `Number: ${number} — Odd Number\n`;
        }

        // 6 & 7) Day using switch statement + weekend check (no prompt)
        let day = "Saturday"; // change this to test different days

        switch (day.toLowerCase()) {
            case "monday":
                output.innerText += "Start of the week!\n";
                break;
            case "tuesday":
                output.innerText += "Keep going strong!\n";
                break;
            case "wednesday":
                output.innerText += "Midweek motivation!\n";
                break;
            case "thursday":
                output.innerText += "Almost there!\n";
                break;
            case "friday":
                output.innerText += "It’s Friday!\n";
                break;
            case "saturday":
            case "sunday":
                output.innerText += "Weekend!\n";
                break;
            default:
                output.innerText += "That’s not a valid day.\n";
                break;
        }

        // 8) loggedIn boolean
        let loggedIn = true;
        if (loggedIn) {
            output.innerText += "Welcome back again.!\n";
        } else {
            output.innerText += "Please log in first.\n";
        }

        // 9) Weather condition
        let weather = "sunny";
        if (weather === "raining") {
            output.innerText += "It's raining.\n";
        } else {
            output.innerText += "Get some nature time, today is a great day!\n";
        }

    })();
</script>
    
```
